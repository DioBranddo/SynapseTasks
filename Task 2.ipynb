### The previous commands were just to make sure we have a workable dataframe now, we'll do some manipulations on the data.

Find out the top 5 most voted candidates in Telangana

```python
df[df['State'] == 'Telangana'].nlargest(5, 'Total Votes')
```

Find out the total number of votes received by each state

```python
totalStateVotes = df.groupby('State')['Total Votes'].sum().reset_index()
totalStateVotes
```

Show these values as a percentage of the total number of votes received in the election.

```python
totalVotes = df['Total Votes'].sum()
```

Now, display the top 5 parties and find out the number of votes received by them

```python
df.groupby('Party')['Total Votes'].sum().reset_index().nlargest(5, 'Total Votes')
```

From the candidates who lost, list 10 candidates who got the highest percent votes.

```python
df[df['Result'] == 'Lost'].groupby('Candidate')['Total Votes'].sum().reset_index().nlargest(10, 'Total Votes')
```

Make a piechart to show the votes recieved by the top 5 candidates in the elections.

```python
topCandidates = df.groupby('Candidate')['Total Votes'].sum().reset_index().nlargest(5, 'Total Votes')
topCandidates.set_index('Candidate')['Total Votes'].plot(kind='pie', autopct='%1.1f%%')

plt.title('Votes Distribution by State')
plt.show()
```

```python
df.info()
```

```python
df['Postal Votes'] = pd.to_numeric(df['Postal Votes'], errors='coerce') #since Postal Votes is of type object, without this it will give eror saying 'no numeric data to plot'
karnataka_pvotes = df[df['State'] == 'Karnataka'].groupby('Party')['Postal Votes'].sum().reset_index()

karnataka_pvotes.plot(kind='bar', x='Party', y='Postal Votes', legend=False)

plt.title('Postal Votes of Parties in Karnataka')
plt.xlabel('Party')
plt.ylabel('Postal Votes')
#plt.yscale('log') #to make the all the bars in the graph clearly visible
plt.xticks(rotation=90, ha='right') 

# Show the plot
plt.show()
```
